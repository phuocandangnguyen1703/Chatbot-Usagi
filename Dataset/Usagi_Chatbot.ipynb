````python
import pandas as pd
from pyvis.network import Network
from IPython.core.display import display, HTML

file = f"/content/drive/MyDrive/Colab Notebooks/ChatBotDataset/Chatbot_new/Dataset/facts.csv"
df = pd.read_csv(file, sep=',', comment='#')

nw = Network(height='1200px', width='100%', bgcolor='white',
             directed=True, notebook=True, cdn_resources='remote')
nw.force_atlas_2based(spring_length=100)

node_size = 16

#Những node nào cùng một group thì vẽ cùng một màu (headtype và tail_type cùng group)
for index, row in df.iterrows():
    src = row['head']
    dst = row['tail']
    edge_label = row['predicate']
    nw.add_node(src, size=node_size, group=row['head_type'])
    nw.add_node(dst, size=node_size, group=row['tail_type'])
    nw.add_edge(src, dst, label=edge_label)

nw.set_edge_smooth('dynamic')
nw.show('kg.html')
display(HTML('kg.html'))
````

````python
'''
Creating KDTree from questions
Vector = question
Input:
    - questions-answers.tsv: contains the list of pairs of question and answer
Output:
    - answers.bin: contains the list of answers
    - kdtree.bin : contains the KD-Tree
'''
import dill as pickle
from scipy.spatial import KDTree
from sentence_transformers import SentenceTransformer
import pandas as pd

# model = SentenceTransformer('all-MiniLM-L6-v2') # Dimensions: 384
model = SentenceTransformer('all-mpnet-base-v2') # Dimensions: 768

path = "/content/drive/MyDrive/Colab Notebooks/ChatBotDataset/Chatbot_new/Dataset"
input_file  = f"{path}/questions-answers.csv"
question_file = f"{path}/questions.bin"
answer_file = f"{path}/answers.bin"
kdtree_file = f"{path}/kdtree.bin"

# [mpid, question, count, answer]
df = pd.read_csv(input_file, sep=',', comment='#')


print("Converting questions to vectors...")
count = 0
questions = []
answers = []
vectors = []
#Xử lý phân tán mới làm dc chỗ này
for index, row in df.iterrows():
    question = row['question']
    answer = row['answer']
    vector = model.encode(question)
    questions.append(question)
    answers.append(answer)
    vectors.append(vector)

print("Creating KDTree...")
kdtree = KDTree(vectors, leafsize=100)
print("Storing questions...")
with open(question_file,'wb') as outfile:
    pickle.dump(questions, outfile)
print("Storing answers...")
with open(answer_file,'wb') as outfile:
    pickle.dump(answers, outfile)
print("Storing KDTree...")
with open(kdtree_file,'wb') as outfile:
    pickle.dump(kdtree, outfile)

print("This task has finished.")
````

````python
import json
import dill as pickle
# import sys
# import threading
import time
from scipy.spatial import KDTree
from sentence_transformers import SentenceTransformer
# loading data

model_name = 'all-mpnet-base-v2' # Dimensions: 768

# specifying data
path = "/content/drive/MyDrive/Colab Notebooks/ChatBotDataset/Chatbot_new/Dataset"
question_file = f"{path}/questions.bin"
answer_file = f"{path}/answers.bin"
kdtree_file = f"{path}/kdtree.bin"

# loading data
print("Loading the Sentence Transformer...")
model = SentenceTransformer(model_name)
with open(kdtree_file, 'rb') as infile:
    kdtree: KDTree = pickle.load(infile)
with open(answer_file, 'rb') as infile:
    answers = pickle.load(infile)
with open(question_file, 'rb') as infile:
    questions = pickle.load(infile)

def chatbot(question):
    if question == '':
        messages = "Bạn còn câu hỏi nào khác không?"
        return json.dumps({"messages": messages}, ensure_ascii=False).encode('utf-8')
    if question == 'exits' or question == 'thoát':
        messages = "Hẹn gặp lại!"
        return json.dumps({"messages": messages}, ensure_ascii=False).encode('utf-8')
    try:
        # convert the question to vector
        vector = model.encode(question)
        # executing the KD-Tree(vector,5)
        dists, idxes = kdtree.query(vector, k=5)
        # get the answer
        answer = answers[idxes[0]]
        print(answer)
        return json.dumps({"messages": answer}, ensure_ascii=False).encode('utf-8')
    except IndexError:
        answer = "Xin lỗi! Chúng tôi chưa hiểu câu hỏi của bạn"
        return json.dumps({"messages": answer}, ensure_ascii=False).encode('utf-8')

question = input("Hãy nhập câu hỏi: ")
response = chatbot(question)
print(response.decode('utf-8'))
````
